```python
import seaborn as sns
#sns.set(color_codes=True)
sns.set()
```

```python
country_counts = df['country'].value_counts()
top_countries = country_counts.head(10)
```

```python
sns.set_style("whitegrid")

plt.Figure(figsize=(10,6))

plt.bar(
    top_countries.index,
    top_countries.values,
    width=0.7,
    alpha=0.8,
    color="steelblue",
    edgecolor="black"
)

plt.title("Most Visited Countries", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Visited", fontsize=12)

plt.ylim(0, top_countries.values.max() * 1.1)
plt.xlim(-0.5, len(top_countries) - 0.5)

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

Q4: Draw a  matplotlib and seaborn line chart from countdata. 

Please style the chart to make it look nicer with good size and alpha settings. 

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))

plt.plot(
    countdata.index,
    countdata.values,
    alpha=0.8,
    linewidth=2
)

plt.title("Trend of Counts Over Time", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Count", fontsize=15)

plt.tight_layout()
plt.show()
```

Q5: Now you try: re-create the plot above with different line widths, colors, and point markers. 

Play with moving the annotation labels around. Make the vertical tick marks 3 days apart instead of 5.

experiment with a more advanced challenge like smoothing the line:

- https://stackoverflow.com/questions/5283649/plot-smooth-line-with-pyplot

- https://stackoverflow.com/questions/25825946/generating-smooth-line-graph-using-matplotlib

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))

plt.plot(
    countdata.index,
    countdata.values,
    color="steelblue",
    marker="o",
    markersize=4,
    alpha=0.6,
    label="Original data",
    linewidth=2
)

smoothed = countdata.rolling(window=5).mean()

plt.plot(
    smoothed.index,
    smoothed.values,
    linewidth=3,
    color="darkred",
    alpha=0.9,
    label="Smoothed (5-day rolling mean)"
)


plt.title("Counts Over Time with Styling and Smoothing", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Count", fontsize=12)

tick_positions = countdata.index[::3]
plt.xticks(tick_positions, rotation=45)

max_date = countdata.idxmax()
max_value = countdata.max()

plt.annotate(
    "Peak value",
    xy=(max_date, max_value),
    xytext=(max_date, max_value + max_value * 0.15),
    arrowprops=dict(arrowstyle="->"),
    fontsize=10
)

plt.legend()
plt.tight_layout()
plt.show()
```

# Pie charts

```python
cities = df['city'].value_counts()
print(cities.count())
cities.head()
```

Q6: draw a pie chart for cities.

Please style the chart to make it look nicer. 

```python
plt.figure(figsize=(8,8))

plt.pie(
    cities.values,
    labels=cities.index,
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False,
    wedgeprops={"edgecolor": "white"}
)

plt.title("Most Visited Cited", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
top_cities = df['city'].value_counts().head(6)

other = df['city'].value_counts().iloc[6:].sum()

city_pie = pd.concat([top_cities, pd.Series({'Other': other})])
```

```python
plt.figure(figsize=(8, 8))

plt.pie(
    city_pie.values,
    labels=city_pie.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

plt.title("Most Visited Cities", fontsize=14)
plt.tight_layout()
plt.show()
```
