```python
import math
import random
```

```python
criterion = float(input("Enter convergence criterion: "))
max_draws = int(input("Enter maximum number of draws: "))

inside_circle = 0
total_draws = 0
percent = float("inf")
pi = math.pi

while percent > criterion and total_draws < max_draws :
    x = random.random()
    y = random.random()
    total_draws += 1

    if x**2 + y**2 <=1:
        inside_circle += 1
    
    ratio = 4*inside_circle / total_draws
    percent = abs(ratio - pi) / pi * 100

if total_draws >= max_draws:
    print("Draw limit was exceeded.")
    print(f"Number of draws: {total_draws}")
    print(f"Estimated Pi: {ratio}")
    print(f"Percentage difference: {percent:.6f}%")
else:
    print("Convergence criterion met.")
    print(f"Number of draws: {total_draws}")
    print(f"Estimated Pi: {ratio}")
    print(f"Percentage difference: {percent:.6f}%")
```
