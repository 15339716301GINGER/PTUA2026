3. In some languages every statement ends with a semi-colon, ;. What happens if you put a semi-colon at the end of a Python statement?

```python
a = "hello";
b = 123;
print(a);
print(b)
```

I don't think adding a semicolon at the end of a Python statement has much of an impact.

4. What if you put a period at the end of a statement?

```python
a = "hello".
b = 123.
print(a).
print(b)
```

This will obviously cause an error; Python will recognize the presence of a dot (.).

5. In math notation you can multiply x and y like this: x y. What happens if you try that in Python?

```python
x = 5
y = 2
print(x y)
print(x * y)
```

```python
i = xy
print(i)
```

Clearly, x multiplied by y cannot be expressed in the form of x y.
