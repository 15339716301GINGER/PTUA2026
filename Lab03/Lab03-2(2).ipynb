```python
import math
import random
import statistics
```

```python
def estimate_pi(criterion):
    inside_circle = 0
    total_draws = 0
    percent = float("inf")
    pi = math.pi

    while percent > criterion:
        x = random.random()
        y = random.random()
        total_draws += 1

        if x**2 + y**2 <=1:
            inside_circle += 1
        ratio = 4 * inside_circle / total_draws
        percent = abs(ratio - pi) / pi * 100
    return total_draws
```

```python
criteria = [0.01, 0.001, 0.0001, 0.00001]

for criterion in criteria:
    draws_required = []
    
    for _ in range(10):
        draws = estimate_pi(criterion)
        draws_required.append(draws)
    
    avg_draws = statistics.mean(draws_required)
    std_draws = statistics.stdev(draws_required)
    
    print(f"Convergence criterion: {criterion}%")
    print(f"Average number of draws: {avg_draws:.2f}")
    print(f"Standard deviation: {std_draws:.2f}")
    print("-" * 40)
```
